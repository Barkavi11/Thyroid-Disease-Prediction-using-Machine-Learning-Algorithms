```python
df.drop(columns=['TSH_measured', 'T3_measured', 'TT4_measured', 'T4U_measured', 'FTI_measured', 'TBG_measured', 'patient_id', 'TBG'], inplace=True)
df.columns
```

```python
col_with_null = ['TSH', 'T3', 'TT4', 'T4U', 'FTI']
for col in col_with_null:
    df[col] = df[col].fillna(0)
```
